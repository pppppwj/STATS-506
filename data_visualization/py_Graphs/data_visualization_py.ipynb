```python
import matplotlib.pyplot as plt
#visualize part
#input data
import pandas as pd
data_vis = pd.read_csv("/Users/weijiepan/Desktop/506_project/506_project_visualization_data.csv")
data_vis = data_vis.drop(["Unnamed: 0"],axis=1)
```

```python
#encode the variables
data_label = data_vis
#younger teenager adult seniors elder
#age_dict = {1:"younger",2:"teenager",3:"adult",4:"seniors",5:"elder"}
data_label.loc[data_label["age"]==1,"age"]="younger 0-12"
data_label.loc[data_label["age"]==2,"age"]="teenager 13-18"
data_label.loc[data_label["age"]==3,"age"]="adult 19-40"
data_label.loc[data_label["age"]==4,"age"]="seniors 41-59"
data_label.loc[data_label["age"]==5,"age"]="elder 59+"

#gender
#gender_dict = {1:"male",2:"female"}
data_label.loc[data_label["gender"]==1,"gender"]="male"
data_label.loc[data_label["gender"]==2,"gender"]="female"

#insurance
data_label.loc[data_label["insurance"]==1,"insurance"]="With_ins"
data_label.loc[data_label["insurance"]==0,"insurance"]="Without_ins"
```

```python
#VA plot
VA = data_label.groupby(["diabetes","age","gender","insurance"])["VA"].mean()   
plt.rcParams['figure.figsize'] = (8.0, 10.0)
VA.unstack(0).plot.barh(title="Vitamin A")
```

```python
#VC plot
VC = data_label.groupby(["diabetes","age","gender","insurance"])["VC"].mean()  
plt.rcParams['figure.figsize'] = (8.0, 10.0)
VC.unstack(0).plot.barh(title = "Vitamin C")
```

```python
#VE plot
VE = data_label.groupby(["diabetes","age","gender","insurance"])["VE"].mean()   
plt.rcParams['figure.figsize'] = (8.0, 10.0)
VE.unstack(0).plot.barh(title = "Vitamin E")
```
